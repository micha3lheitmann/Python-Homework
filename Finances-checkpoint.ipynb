```python
import pandas as pd
```

```python
Finances = pd.read_csv("Finances.csv")
```

```python
def color_negative_red(value):
  """
  Colors elements in a dateframe
  green if positive and red if
  negative. Does not color NaN
  values.
  """

  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'green'
  else:
    color = 'black'

  return 'color: %s' % color
```

```python
Finances.style.applymap(color_negative_red, subset=['Profit/Losses'])
```
